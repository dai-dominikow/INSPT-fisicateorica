```python
import numpy as np
import matplotlib.pyplot as plt
import math

dt = 0.01
tf = 40
t = np.arange(0 , tf + dt ,dt)

i=0
theta = np.zeros(len(t))
theta[i] = 0.1
omega = np.zeros(len(t))
omega[i] = 0.1
g = -9.8
l = 2
t_len= len(t)-1
```

```python
i=0

while i < t_len:
    omega[i+1] = omega[i] + ( g/l*math.sin(theta[i])) * dt
    theta[i+1] = theta[i] + omega[i+1] * dt
    i = i+1
```

```python
plt.figure(1)
plt.plot(t,theta,"r")
plt.figure(2)
plt.plot(theta,omega,"b")
```
